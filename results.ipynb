```python
import pickle

import numpy as np
import matplotlib.pyplot as plt
```

```python
ckpt_path_list = [
    ".ckpts/pg/CartPole-v1-0/",
    ".ckpts/pg/CartPole-v1-1/"
]
label_list = [
    "Policy-Gradient",
    "Policy-Gradient with normalizing returns"
]

plt.figure(figsize=[15, 5], facecolor="white")
for i, ckpt_path in enumerate(ckpt_path_list):
    results = []
    for j in range(3):
        with open(ckpt_path + "results_%i.pkl" % j, "rb") as f:
            results.append(pickle.load(f))
    
    results = np.array(results)
    
    plt.plot(range(1, results.shape[1] + 1), np.mean(results, axis=0), label=label_list[i])
    plt.fill_between(
        range(1, results.shape[1] + 1),
        np.mean(results, axis=0) - np.std(results, axis=0),
        np.mean(results, axis=0) + np.std(results, axis=0),
        alpha=.2
    )

plt.grid()
plt.legend(loc='best')
plt.show()
```
